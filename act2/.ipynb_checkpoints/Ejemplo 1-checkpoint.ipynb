```python
# Code source: GaÃ«l Varoquaux
# Modified for documentation by Jaques Grobler
# License: BSD 3 clause

import matplotlib
import matplotlib.pyplot as plt

import mpl_toolkits.mplot3d  # noqa: F401
import numpy as np

from sklearn import datasets
from sklearn.cluster import KMeans

np.random.seed(5)

dataSetEscogido = datasets.load_diabetes()
X = dataSetEscogido.data
y = dataSetEscogido.target

estimators = [
 ("k means diabetes 8", KMeans(n_clusters=8)),
 ("k means diabetes 3", KMeans(n_clusters=3)),
 ("k means diabetes bad_init", KMeans(n_clusters=3, n_init=1, init="random")),
 ("K means diabetes k=5", KMeans(n_clusters=5)),
]
```

```python
fig = plt.figure(figsize=(10, 8))
for idx, (name, est) in enumerate(estimators):
    ax = fig.add_subplot(2, 2, idx + 1, projection="3d", elev=48, azim=134)
    est.fit(X)
    labels = est.labels_

    ax.scatter(X[:, 3], X[:, 0], X[:, 2], c=labels.astype(float), edgecolor="k")

    ax.xaxis.set_ticklabels([])
    ax.yaxis.set_ticklabels([])
    ax.zaxis.set_ticklabels([])
    ax.set_title(name)
```
