## Hold out

El hold out specifica que porcentaje del dataset se va a dejar afuera para validaciÃ³n/testing. en este caso un r, donde r pertenece a [0,1]

```python
def holdOut(li, r): # Recibe un listado iterable que contenga los filas o instancias
    # Logra separar en classes discretas
    l = li.drop_duplicates(ignore_index=True)
    tam = len(l.index)
    elems = list(range(tam))
    clases = {}
    for a in range(tam):
        # print(l.iat[a, -1])  # Se asume que la clase esta en la ultima columna
        c = l.iat[a,-1]
        if c not in clases:
            clases[c] = []
        else:
            clases[c].append(a)
    ent = []
    val = []
    for k in clases:
        t = len(clases[k])
        v = int(t*r)
        shuffle(clases[k])
        val += clases[k][0:v]
        ent += clases[k][v:]
    return l.iloc[ent],l.iloc[val]

ent,val = holdOut(union, 0.3)
```

```python
def elemComun(a, b): # a y b son datasets de pandas, retorna true si existen elementos en comun
    tamA = len(a.index)
    tamB = len(b.index)
    df3 = pd.concat([a,b], ignore_index=True)
    df3 = df3.drop_duplicates(ignore_index=True)
    tamRes = len(df3.index)
    print(tamA,tamB,tamRes)
    if((tamA + tamB) > tamRes):
        return True # Existen elementos en comun
    else:
        return False # No existen elementos en comun
    
```

## K-fold cross-validation

No estoy seguro de como recibir una funciÃ³n de predicciÃ³n, entonces va a retornar k tuples, donde el primer conjunto es el entrenamiento, y el segundo los resultados

```python
def kfold(f, k):
    frame = f.drop_duplicates(ignore_index=True)
    nums = list(range(len(frame.index)))
    shuffle(nums)
    # El ultimo conjunto tendra mas elementos de ser el caso de sobra
    part = []
    div = int((len(frame.index))/k)
    # Particionamos el frame
    for a in range(k-1):
        part.append(frame.iloc[nums[a*div:(a+1)*div]])
    part.append(frame.iloc[nums[(k-1)*div:]])
    # Part tiene las particiones en k
    # Ahora realizamos los tuples
    ret =[]
    for a in range(k):
        ret.append( (pd.concat(part[:a] + part[a+1:]),part[a]) )
    return ret
```

```python
res = kfold(union, 3)
for a in res:
    #print("--")
    #print(a[0])
    #print(a[1])
    #print("---")
    print(len(a[0].index), len(a[1].index))
    print(elemComun(a[0], a[1]))
res2 = kfold(union2, 10)
for a in res2:
    print(elemComun(a[0], a[1]))
```

```python
def leaveOneOut(f):
    res = []
    frame = f.drop_duplicates(ignore_index=True)
    tam = len(frame.index)
    for a in range(tam):
        r = list(range(tam))
        del r[a]
        res.append( (frame.iloc[r], frame.iloc[a]) )
    return res
```

## Algoritmos de aprendizaje

### 1nn

```python
def contieneNull(a):
    return (a.isnull().values.any())

def nnClass(data, fila):
    d = data.reset_index()
    l = list(d.columns.values)[1:-1] # Excluye el ultimo columna
    df = pd.DataFrame()
    df['acum'] = [0]*len(d)
    for a in l:
        df['acum'] = (d[a]-fila[a])*(d[a]-fila[a])
    df['acum'] = np.sqrt(df['acum'])
    i=df['acum'].idxmin()
    return (d.iloc[i][list(d.columns.values)[-1]])
```

### Distancia mÃ­nima

```python
def disMinEnt(data): # Retorna el matriz de los centroides
    a = data.groupby(data.columns[-1]).mean().reset_index()
    elems = list(a.columns)
    a = a[elems[1:] + [elems[0]]] # Movemos los clases al final
    return a
```

## Accuracy y matriz de confusion

```python
def matEsp(pares, e): # Realiza el matriz y el accuracy sobre un solo valor
    tp = 0 # true negative
    fn = 0 # false negative
    fp = 0 # false positive
    tn = 0 # true negative
    for a in pares:
        if a[0]==e:
            # Realmente es positivo
            if a[1]==e:
                tp+=1
            else:
                fn+=1
        else:
             if a[1]==e:
                 fp+=1
             else:
                 tn+=1
    print(f"Elaborado sobre {e}")
    print(f"Accuracy: {(tp+tn)/(tp+tn+fp+fn)}")
    print("\t1\t0")
    print(f"1\t{tp}\t{fn}")
    print(f"0\t{fp}\t{tn}")
            
    
def accYMat(pares): # Sobre los pares de (real,prediccion), se obtiene el accuracy y el matriz de confusion
    s = set()
    for a in pares:
        s.add(a[0])
    print("Arriba es la clase predicha, a la izquierda es la clase real.")
    for a in s:
        matEsp(pares,a)
```

### ImplementaciÃ³n de algoritmos de aprendizaje con algoritmos de validaciÃ³n

```python
def nnHoldOut(data):
    ent,val = holdOut(data, 0.3)
    # Calculamos un vector de distancia
    real = []
    pred = []
    for a in range(len(val)):
        real.append(val.iloc[a][val.columns[-1]])
        pred.append(nnClass(ent, val.iloc[a]))
    return list(zip(real,pred))
def nnKFold(data):
    k = kfold(data, 10)
    #ent,val = holdOut(data, 0.3)
    # Calculamos un vector de distancia
    real = []
    pred = []
    for b in k:
        ent,val = b
        for a in range(len(val)):
            real.append(val.iloc[a][val.columns[-1]])
            pred.append(nnClass(ent, val.iloc[a]))
    return list(zip(real,pred))
def nnOneOut(f):
    # res = []
    real = []
    pred = []

    frame = f.drop_duplicates(ignore_index=True)
    tam = len(frame.index)
    for a in range(tam):
        r = list(range(tam))
        del r[a]
        ent = frame.iloc[r]
        val = frame.iloc[a]
        real.append(val[ent.columns[-1]])
        pred.append(nnClass(ent, val))
    return list(zip(real,pred))
```

```python
def distHoldOut(data):
    e,val = holdOut(data, 0.3)
    ent = disMinEnt(e)
    # Calculamos un vector de distancia
    real = []
    pred = []
    for a in range(len(val)):
        real.append(val.iloc[a][val.columns[-1]])
        pred.append(nnClass(ent, val.iloc[a]))
    return list(zip(real,pred))
def distKFold(data):
    k = kfold(data, 10)
    #ent,val = holdOut(data, 0.3)
    # Calculamos un vector de distancia
    real = []
    pred = []
    for b in k:
        e,val = b
        ent = disMinEnt(e)
        for a in range(len(val)):
            real.append(val.iloc[a][val.columns[-1]])
            pred.append(nnClass(ent, val.iloc[a]))
    return list(zip(real,pred))
def distOneOut(f):
    # res = []
    real = []
    pred = []

    frame = f.drop_duplicates(ignore_index=True)
    tam = len(frame.index)
    for a in range(tam):
        r = list(range(tam))
        del r[a]
        e = frame.iloc[r]
        ent = disMinEnt(e)
        val = frame.iloc[a]
        real.append(val[ent.columns[-1]])
        pred.append(nnClass(ent, val))
    return list(zip(real,pred))
```
